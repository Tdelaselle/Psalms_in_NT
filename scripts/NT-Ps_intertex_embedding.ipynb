# NT-Ps Intertext embedding from distance matrices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from tqdm import tqdm
```

## Load data and distances matrices

```python
Ps = pickle.load(open("data/Ps.pkl", "rb"))
NT = pickle.load(open("data/NT.pkl", "rb"))
Dn1 = pickle.load(open("distance_matrices/Ps_NT_norm1.pkl", "rb"))
Dn2 = pickle.load(open("distance_matrices/Ps_NT_norm2.pkl", "rb"))
```

## Form NT-Ps measures dataset

```python
sizes = Dn1['token'].shape
n_features = 16

Nt_Ps = np.zeros((sizes[0]*sizes[1],n_features),dtype=object)
print(Nt_Ps.shape)
Nt_Ps[:,0] = Dn1['token'].flatten(order='F')
Nt_Ps[:,1] = Dn1['lemma'].flatten(order='F')
Nt_Ps[:,2] = Dn1['pos'].flatten(order='F')
Nt_Ps[:,3] = Dn1['stop_words'].flatten(order='F')
Nt_Ps[:,4] = Dn1['lemma_filtered'].flatten(order='F')
Nt_Ps[:,5] = Dn1['lexicon'].flatten(order='F')
Nt_Ps[:,6] = Dn1['domain'].flatten(order='F')
Nt_Ps[:,7] = Dn1['subdomain'].flatten(order='F')
del Dn1
Nt_Ps[:,8] = Dn2['token'].flatten(order='F')
Nt_Ps[:,9] = Dn2['lemma'].flatten(order='F')
Nt_Ps[:,10] = Dn2['pos'].flatten(order='F')
Nt_Ps[:,11] = Dn2['stop_words'].flatten(order='F')
Nt_Ps[:,12] = Dn2['lemma_filtered'].flatten(order='F')
Nt_Ps[:,13] = Dn2['lexicon'].flatten(order='F')
Nt_Ps[:,14] = Dn2['domain'].flatten(order='F')
Nt_Ps[:,15] = Dn2['subdomain'].flatten(order='F')
del Dn2
```

```python
NT_Ps_verse = pd.DataFrame(columns=['NT','NT_part','Ps','Ps_part'])

# Nt and Ps verses names
NT_Ps_verse.loc[:,'NT'] = [i for i in NT['verse'] for j in range(sizes[0])]
NT_Ps_verse.loc[:,'NT_part'] = [i for i in NT['verse_part'] for j in range(sizes[0])]
NT_Ps_verse['Ps'] = [j for _ in range(sizes[1]) for j in Ps['verse']]
NT_Ps_verse['Ps_part'] = [j for _ in range(sizes[1]) for j in Ps['verse_part']]
NT_Ps_verse['NT_token_len'] = [len(i) for i in NT['token'] for j in range(sizes[0])]
NT_Ps_verse['NT_lexic_len'] = [len(i) for i in NT['lexicon'] for j in range(sizes[0])]
NT_Ps_verse['Ps_token_len'] = [len(j) for _ in range(sizes[1]) for j in Ps['token']]
NT_Ps_verse['Ps_lexic_len'] = [len(j) for _ in range(sizes[1]) for j in Ps['lexicon']]
display(NT_Ps_verse)
```

```python
Nt_Ps = pd.DataFrame(Nt_Ps)
Nt_Ps = Nt_Ps.astype(float)
Nt_Ps = pd.concat([NT_Ps_verse, Nt_Ps], axis=1)
embedding = Nt_Ps.rename(columns={0: 'token 1', 1: 'lemma 1', 2: 'PoS 1', 3: 'stop_words 1', 4: 'lemma_filtered 1', 5: 'lexicon 1', 6: 'domain 1', 7:'subdomain 1', 8: 'token 2', 9: 'lemma 2', 10: 'PoS 2', 11: 'stop_words 2', 12: 'lemma_filtered 2', 13: 'lexicon 2', 14: 'domain 2', 15:'subdomain 2'})
del NT_Ps_verse, Nt_Ps
display(embedding.head(30))
```

```python
feat_1 = 8 # position of the first feature column in the dataframe

# handle inf in all columns by replacing them with the mean of the column
for col in tqdm(embedding.columns[feat_1:(feat_1+n_features)]):
    ids = np.isinf(embedding[col])
    mean_value = embedding.loc[~ids, col].mean()
    embedding.loc[ids, col] = mean_value
    print(col, " : ", ids.sum(), " inf values replaced")
# handle NaN in all columns by replacing them with the mean of the column
for col in tqdm(embedding.columns[feat_1:(feat_1+n_features)]):
    ids = np.isnan(embedding[col])
    mean_value = embedding.loc[~ids, col].mean()
    embedding.loc[ids, col] = mean_value
    print(col, " : ", ids.sum(), " NaN values replaced")
```

## Saving intertext embedding

```python
with open("distance_matrices/NT-Ps_embedding.pkl", "wb") as file:
    pickle.dump(embedding, file, protocol=pickle.HIGHEST_PROTOCOL)
```
